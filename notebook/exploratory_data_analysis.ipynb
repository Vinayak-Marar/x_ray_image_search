```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
file_path = "../data/raw"
```

```python
from PIL import Image
import os
```

```python
import random
```

```python
dirs = ["chest","dental","hand"]
```

```python
for folder in dirs:
    
    files = os.listdir(os.path.join(file_path,folder))
    print(folder," ",len(files))
    
    f = random.choice(files)
    f = os.path.join(os.path.join(file_path,folder),f)

    img = Image.open(f)


    plt.imshow(img,cmap="gray")
    plt.show()
```

```python
from collections import Counter
```

```python

for folder in dirs:
    sizes = []
    files = os.listdir(os.path.join(file_path,folder))

    for f in files:
        full_path = os.path.join(os.path.join(file_path,folder),f)
        img = Image.open(full_path)
        sizes.append(img.size)

    print(folder," ",Counter(sizes).most_common(5))
```
